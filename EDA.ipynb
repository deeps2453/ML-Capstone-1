```python
import pandas as pd
import matplotlib.pyplot as plt

# load your dataset
df = pd.read_csv("wages.csv")


numeric_cols = [
    "Caps",
    "Apps",
    "Age",
    "Reputation",
    "Salary"
]

for col in numeric_cols:
    plt.figure()
    df[col].hist(bins=30)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
sns.histplot(df.Caps[(df.Caps >= 10) & (df.Caps <= 100)], bins=50)
```

```python
sns.histplot(df.Apps[(df.Apps >= 100) & (df.Apps <= 800)], bins=50)
```

```python
sns.histplot(df.Apps[(df.Apps >= 590) & (df.Apps <= 800)], bins=50)
```

```python
sns.histplot(df.Salary[(df.Salary >= 100000) & (df.Salary <= 800000)], bins=50)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("wages.csv")

# select only numeric columns (heatmaps require numeric matrix)
num_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(12,8))
sns.heatmap(num_df.corr(), annot=True, cmap="viridis")
plt.show()
```
